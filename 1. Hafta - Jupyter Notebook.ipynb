#Gerekli Paketler

```python
import numpy as np
import matplotlib.pyplot as plt

from skimage import io, color, data
from skimage.transform import resize
```

# Görüntü Okuma & Gösterme (scikit-image kütüphanesi)

Kütüphanedeki diğer içeriklere [Linkten](http://scikit-image.org/docs/stable/api/skimage.data.html) ulaşabilirsiniz.




```python
# Kütüphanedeki hazır datasetlerden örnek bir görüntüyü çağır.
img = data.cat()
# "imshow" komutu görüntüyü göstermek için kullanılır.
plt.imshow(img)
plt.axis("off")
plt.show()
```

# Görüntü Matrisi

```python
# Görüntü matrisinin boyutları
print("shape:", img.shape)
# Görüntü matrisindeki piksellerin veri tipi
print("dtype:", img.dtype)
# 100. satır, 200. sütundaki pikselin değerini göster.
print("Bir piksel (y=100, x=200):", img[100, 200])  # RGB ise 3 değer döner
```

# Gri Seviyeye Dönüştürme

```python
gray = color.rgb2gray(img)
plt.imshow(gray, cmap="gray")
plt.axis("off")
plt.show()

print("gray shape:", gray.shape)
print("gray dtype:", gray.dtype)
```

# Boyut değiştirme

```python
small = resize(img, (200, 200), anti_aliasing=True)

plt.imshow(small)
plt.axis("off")
plt.show()
```

# Kırpma

```python
crop = img[50:250, 50:250]
plt.imshow(crop)
plt.axis("off")
plt.show()
```
